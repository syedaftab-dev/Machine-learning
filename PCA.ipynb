## PRINCIPLE COMPONENT ANALYSISI

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# loading the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
cancer_dataset.DESCR
```

data->key consists of all data and feature_names will have all the feauture list

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

# applying PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # converts all features to 2 features with max varinace
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_ # variance of featur1 and 2
```

# Lets Visualize

```python
plt.figure(figsize=(8,6))
plt.scatter(x=data_pca[:,0],y=data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
