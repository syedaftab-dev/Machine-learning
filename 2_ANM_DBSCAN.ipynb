# Anamoly detection in DBSCAN clustering

```python
from sklearn.cluster import DBSCAN
import numpy as np
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbscan=DBSCAN(eps=0.10) # eps->epsilon circle radius
```

```python
dbscan.fit(X)
```

```python
dbscan.fit_predict(X)
```

```python
dbscan.labels_
#he dbscan.labels_ attribute in scikit-learn's DBSCAN implementation returns an array containing the cluster labels for each point in the dataset that was used to fit the model
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
#purple are out;iers
```
