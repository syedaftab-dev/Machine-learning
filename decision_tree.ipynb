# **Decision tree implementation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

## it has 4 features ['sepal lenght','sepal width','petal length','petal width'] and an ouput feature target with 3 categories ['setosa', 'versicolor', 'virginica']

## so based on feature we have determine which flower is it

---



```python
print(iris['DESCR']) # info of data
```

```python
# independent and dependent features

X=pd.DataFrame(iris['data'],columns=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
```

```python
y=iris['target']
```

```python
from os import X_OK
# train test split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# apply decision tree classifier

from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True) # filled-> with values
```

## in this overfitting is seen so we use pruning

if we set max_depth=2 inside DecisionTreeClassifier(max_depth=2) this will only give tree till depth 2

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# 93% accuracy with depth=2 AND 97% full tree
```

### **Decision Tree classifier prepruning and Hyperparameter tining**

```python
params ={
    'criterion':['gini','entropy','log_loss'],
    'splitter': ['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(treeclassifier,param_grid=params,cv=5,scoring='accuracy')
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# accuracy 100%
```
